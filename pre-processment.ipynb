```python
import pandas as pd
import glob
import os
from sklearn import preprocessing
pd.set_option('display.max_columns', None)
os.getcwd()
```

### separating data set atributtes into features and prediction_class 

```python
attributes = [
   "Wife's age",
   "Wife's education",
   "Husband's education",
   "Number of children ever born",
   "Wife's religion",
   "Wife's now working?",
   "Husband's occupation",
   "Standard-of-living index",
   "Media exposure",                
   "Contraceptive method used"
]

features = [
    "Wife's age",
    "Wife's education",
    "Husband's education",
    "Number of children ever born",
    "Wife's religion",
    "Wife's now working?",
    "Husband's occupation",
    "Standard-of-living index",
    "Media exposure"
]
prediction_class = [
    "Contraceptive method used"
]

```

# !! to import data set, convert the downloaded data file into .txt ( by replacing the end of the name of the file into .txt) file and upload it on the same directory that you started jupyter notebook !!

```python
data_set = pd.read_csv(glob.glob(os.getcwd()+'\\*.txt')[0], names = attributes)
```

### normalizing features and class(need to see missing values pre processment)

```python
def normalize_data_features(data_set):
    for key in features:
        min_feature = data_set[key].min()
        max_feature = data_set[key].max()
        data_set['normalized_'+key] = data_set[key].map(lambda row: (float)(row-min_feature)/(min_feature+max_feature))
    return data_set
    
```

```python
#NEED TO SEE MISSING VALUES!!!!!!!
data_set = normalize_data_features(data_set)
```

```python
data_set['normalized_class'] = pd.get_dummies(data_set['Contraceptive method used']).to_dict('split')['data']
```
